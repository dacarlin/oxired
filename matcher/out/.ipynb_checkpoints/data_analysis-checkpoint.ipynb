```python
import pandas
sf = pandas.read_csv('scoreup_to_100.sc', delimiter=r'\s+')
sf.head()
```

```python
# constraint cutoff 
sf = sf[( sf['all_cst'] < 10 )]

# additional columns 
sf['scaffold'] = sf['description'].str[0:4]
sf['mutant'] = sf['description'].str[:-4:]

def top_5(df, quantile):
    '''Returns lowest 5% '''
    return df[(df['total_score'] < df['total_score'].quantile(quantile))].sort('total_score')

sf.groupby(by='scaffold').apply(top_5, quantile=(0.10))
```
